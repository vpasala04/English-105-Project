```python
# Now, we need to import the .csv file. We do this by creating a dataframe object.

# Type the command "pd.read_csv". This allows for the pandas function to read through the .csv file as you're recalling certain aspects of the data later

# To combine these two aspects of the code, we would need to create the dataframe object by typing "df = pd.read_csv("CountyHealthData_2014-2015.csv")". Within the parentheses, you ust need to write in the file name of your csv file. In this repository, we are using the County Health Public Data so that file is what will go into the parentheses.
```

```python
df=pd.read_csv("CountyHealthData_2014-2015.csv")
```

```python
## Indexing to Manipulate a Larger Dataset

# Now, we will work on filtering our very large dataset from the .csv file into certain sections that we want to examine specifically in this repository. A common and quick method is using indexing, where you choose certain columns and rows.


# To do this, we will use the command "df.loc[:,["Name of Desired Column", "Name of Desired Column"]]" 


# An important note is that you can include as many or as little column as you want by separating each name of a column with a comma, as shown above.


# An easy shortcut that we will be using today for rows is using the ".head(n = number of rows you want presented)", where you will get the rows starting from the beginning of the dataset.


# The result after running the code should be the table that you are examining and wanting to analyze.
```

```python
df.loc[:,["State","County","Teen births", "High school graduation"]].head(n=46)
```

```python
TeenCountyData = df.loc[:,["State","County","Teen births", "High school graduation"]].head(n=46).copy()
```

```python
## Creating a New Data File for The "Edited" Data


# As seen above, the new data set was given the name "STICountyData". By simply typing the name of the new data, and run the strip of text, you will get back the new data table that you indexed.
```

```python
TeenCountyData
```

```python
# The last thing to do is now establish this new data table as a separate file. We can do this by typing "Name of New Data.to_csv("Name of New Data", index=false)". This strip of code will transform this cut dataset into its own csv file.
```

```python
STICountyData.to_csv("TeenCountyData.csv", index=False)
```
